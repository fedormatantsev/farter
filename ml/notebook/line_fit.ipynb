```python
import datetime

from acquisition import fetch_candle_data
from processing.plot_utils import make_candlestick
from processing.fit_line import fit_line

import numpy as np
from numpy.random import default_rng

from plotly import graph_objects as go

from ta.trend import SMAIndicator


candle_data = fetch_candle_data(
    ticker='NVDA',
    first_date=datetime.date(year=2021, month=4, day=1),
    last_date=datetime.date(year=2021, month=7, day=24))

sma13 = SMAIndicator(candle_data.c, window=13).sma_indicator().shift(8)
sma8 = SMAIndicator(candle_data.c, window=8).sma_indicator().shift(5)

subrange = candle_data[-43:-38]

a_sma, b_sma = fit_line(sma13[subrange.index], sma8[subrange.index])
a, b = fit_line(subrange.h, subrange.l)

fig = go.Figure()
fig.add_trace(make_candlestick(candle_data))

y_fit = a * subrange.index.values + b
y_sma = a_sma * subrange.index.values + b_sma
fig.add_scatter(x=subrange.t, y=y_fit, name='candles_fit')
fig.add_scatter(x=subrange.t, y=y_sma, name='sma_fit')
fig.add_scatter(x=candle_data.t, y=sma13, name='sma13')
fig.add_scatter(x=candle_data.t, y=sma8, name='sma8')

fig.update_layout(height=800)
fig.show()

alpha = np.arctan(a) * 180 / 3.1415
beta = np.arctan(a_sma) * 180 / 3.1415

print(abs(alpha - beta))
```
